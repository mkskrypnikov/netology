# Задание 1

```python
def com_p (num):
    check = 0
    for i in documents:
        if num == i['number']:
            print('Владелец документа: ' + i['name'])
            check += 1
    if check == 0:
        print('Документ не найден в базе')       
        
```

```python
def com_s(num):
    check = 0
    for i in directories:
        for j in directories[i]:
            if j == num:
                print('Документ хранится на полке: ' + i)
                check += 1
    if check == 0:
        print('Документ не найден в базе')  
        
```

```python
def com_l():
    for i in documents:
        x = '№: ' + i['number'] + ', тип: '+ i['type'] + ', владелец: ' + i['name']
        for j in directories:
            if i['number'] in directories[j]:
                y = ', полка хранения: ' + j
        print(x + y) 

        
```

```python
def com_ads(num):
    if num in directories:
        print('Такая полка уже существует. Текущий перечень полок: ' + str(list(directories))[1:-1])
    else:
        directories.update({num:[]})
        print('Полка добавлена. Текущий перечень полок: ' + str(list(directories))[1:-1])
    
```

```python
def com_ds(num):
    if num not in directories:
            print('Такой полки не существует. Текущий перечень полок: ' + str(list(directories))[1:-1])
    else:
        if len(directories[num]) > 0:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + str(list(directories))[1:-1])
        else:
            directories.pop(num)
            print('Полка удалена. Текущий перечень полок: ' + str(list(directories))[1:-1])
```

```python
choose_com = str(input('Введите команду: ')).lower()

if choose_com == 'p':
    num = str(input('Введите номер документа: ')).lower()
    com_p(num)
elif choose_com == 's':
    num = str(input('Введите номер документа: ')).lower()
    com_s(num)
elif choose_com == 'l':
    com_l()
elif choose_com == 'ads':
    num = str(input('Введите номер полки: ')).lower()
    com_ads(num)
elif choose_com == 'ds':
    num = str(input('Введите номер полки: ')).lower()
    com_ds(num)
    
```

# Задание 2 (необязательное)

```python
def com_ad(doc_dir, doc_type, doc_num, doc_name):
    if doc_dir not in directories:
        print('Такой полки не существует. Добавьте полку командой as:')
        print('Текущий список документов: ')
        com_l()
    else:
        documents.append({"type": doc_type, "number": doc_num, "name": doc_name})
        directories[doc_dir].append(doc_num)
        print('Документ добавлен. Текущий список документов: ')
        com_l()
```

```python
def com_d(num):
    f = ''
    x = 0
    for j in directories:
        for y in directories[j]:
            if y == num:
                f = j
                break
            else:
                f = f
    if f != '':
        for z in (directories[f]):
            if z == num:
                directories[f].remove(z)
        for i in documents:
            if i['number'] == num:

                x = x
                break
            else:
                x += 1
        del documents[x]
        print('Документ удален.','Текущий список документов:', sep = '\n')        
        com_l()
    else:
        print('Документ не найден в базе.','Текущий список документов:', sep = '\n')        
        com_l()



```

```python
def com_m(num, dir_num):
    check_num = 0
    check_dir = 0
    for i in directories:
        if dir_num == i:
            check_dir = 1
        else:
            check_dir = check_dir
        for j in directories[i]:
            if num == j:
                x = i
                y = j
                check_num = 1
            else:
                check_num = check_num
    if check_dir == 1 and check_num == 1:
        directories[dir_num].append(num)
        directories[x].remove(num)
        print('Документ перемещен.', 'Текущий список документов:', sep = '\n')
        com_l()
    elif check_num == 0:
        print('Документ не найден в базе.', 'Текущий список документов:', sep = '\n')
        com_l()
    elif check_dir == 0:
        print('Такой полки не существует. Текущий перечень полок: ' + str(list(directories))[1:-1] )
    
    
```

```python
choose_com = str(input('Введите команду: ')).lower()

if choose_com == 'p':
    num = str(input('Введите номер документа: ')).lower()
    com_p(num)
elif choose_com == 's':
    num = str(input('Введите номер документа: ')).lower()
    com_s(num)
elif choose_com == 'l':
    com_l()
elif choose_com == 'ads':
    num = str(input('Введите номер полки: ')).lower()
    com_ads(num)
elif choose_com == 'ds':
    num = str(input('Введите номер полки: ')).lower()
    com_ds(num)
elif choose_com == 'ad':
    doc_num = str(input('Введите номер документа: ')).lower()
    doc_type = str(input('Введите тип документа: ')).lower()
    doc_name = str(input('Введите владельца документа: ')).lower()
    doc_dir = str(input('Введите полку для хранения: ')).lower()
    com_ad(doc_dir, doc_name, doc_num, doc_type)
elif choose_com == 'd':
    num = str(input('Введите номер документа: ')).lower()
    com_d(num)
elif choose_com == 'm':
    num = str(input('Введите номер документа: ')).lower()
    dir_num = str(input('Введите номер полки: ')).lower()
    com_m(num, dir_num)
```
