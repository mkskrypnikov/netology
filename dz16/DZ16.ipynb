# Задание 1. Базовое изучение
Изучить представленный набор данных на основе описания его столбцов и выбрать 8 столбцов для дальнейшего изучения (среди них должны быть как числовые, так и категориальные). Провести расчет базовых метрик для них, кратко описать результаты.



```python
surgery = {1 : 'yes', 2 : 'no'}

age = {1 : 'adult', 2: 'young', 9 : 'young'}

abdominal_distension = {1 : 'none',
               2 : 'slight',
               3 : 'moderate',
               4 : 'severe',
               5 : 'severe'}
pain = { 1 : 'no pain',
               2 : 'depressed',
               3 : 'intermittent mild pain',
               4 : 'intermittent severe pain',
               5 : 'continuous severe pain'}

abdomen = {1 : 'normal',
               2 : 'other',
               3 : 'firm feces in the large intestine',
               4 : 'distended small intestine',
               5 : 'distended large intestine'}

outcome = {1 : 'lived',
           2 : 'died',
           3 : 'was euthanized'}

site_of_lesion = {1 : 'gastric',
               2 : 'sm intestine',
               3 : 'lg colon',
               4 : 'lg colon and cecum',
               5 : 'cecum',
               6 : 'transverse colon',
               7 : 'retum/descending colon',
               8 : 'uterus',
               9 : 'bladder',
               11 : 'all intestinal sites',
               0 : 'none'}
```

```python
df_horse = pd.read_csv('horse_data.csv', header=None, na_values='?')
df_horse = df_horse[[0, 1, 3, 4,10, 12, 17, 22]]
df_horse.columns = ['surgery',
'Age',
'rectal_temperature',
'pulse',
'abdominal_distension',                    
'pain',
'abdomen',
'outcome'
]
df_horse_outliers = df_horse.copy()

df_horse.head()
```

```python
df_horse.info()
```

```python
df_horse['surgery'] = df_horse['surgery'].replace(np.NaN, -100).astype(int,errors='ignore')
df_horse['pulse'] = df_horse['pulse'].replace(np.NaN, -100).astype(int,errors='ignore')
df_horse['abdominal_distension'] = df_horse['abdominal_distension'].replace(np.NaN, -100).astype(int,errors='ignore')
df_horse['pain'] = df_horse['pain'].replace(np.NaN, -100).astype(int,errors='ignore')
df_horse['abdomen'] = df_horse['abdomen'].replace(np.NaN, -100).astype(int)
df_horse['outcome'] = df_horse['outcome'].replace(np.NaN, -100).astype(int,errors='ignore')
df_horse.head()
```

```python
def replace_temp(row):
    if row['rectal_temperature'] <= 37.6 and  row['rectal_temperature'] >= 0:
        return 'min'
    elif  row['rectal_temperature'] >= 38:
        return 'max'
    elif row['rectal_temperature'] > 37.6 and  row['rectal_temperature'] < 38:
        return 'norm'
    else:
        return row['rectal_temperature']
    
def replace_pulse(row):
    if row['pulse'] >= 20 and  row['pulse'] <= 40:
        return 'good'
    else:
        return 'bad'
```

```python
df_horse = df_horse.replace({"surgery": surgery})
df_horse = df_horse.replace({"Age": age})
df_horse = df_horse.replace({"abdominal_distension": abdominal_distension})
df_horse = df_horse.replace({"pain": pain})
df_horse = df_horse.replace({"abdomen": abdomen})
df_horse = df_horse.replace({"outcome": outcome})
df_horse = df_horse.replace({"site_of_lesion": site_of_lesion})
df_horse['rectal_temperature'] = df_horse.apply(replace_temp, axis=1)
df_horse['pulse'] = df_horse.apply(replace_pulse, axis=1)
df_horse = df_horse.replace(-100, value=np.nan)
df_horse.head()
```

# Задание 3. Работа с пропусками
Рассчитать количество пропусков для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
def replace_abdominal_distension(row):
    if row['abdominal_distension'] == np.NaN:
        if row['surgery'] == 1 and row['outcome'] == 1:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][0]
        elif row['surgery'] == 1 and row['outcome'] == 2:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][1]
        elif  row['surgery'] == 1 and row['outcome'] == 3:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][2]
        elif  row['surgery'] == 2 and row['outcome'] == 1:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][3]
        elif row['surgery'] == 2 and row['outcome'] == 2:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][4]
        elif row['surgery'] == 2 and row['outcome'] == 3:
            return row.groupby(['surgery', 'outcome']).row.median().reset_index()['abdominal_distension'][5]
        else:
            return row['abdominal_distension']
    else:
        return row['abdominal_distension']
    
    
```

```python
df_horse_notna = df_horse_outliers
df_horse_notna = df_horse_notna.drop(['abdomen'], axis=1)
df_horse_notna = df_horse_notna.dropna(thresh=5)
df_horse_notna['rectal_temperature'] = df_horse_notna['rectal_temperature'].replace(np.NaN, df_horse_notna['rectal_temperature'].median()).astype(float,errors='ignore')
df_horse_notna['abdominal_distension'] = df_horse_notna.apply(replace_abdominal_distension, axis=1)
df_horse_notna.info()
```
